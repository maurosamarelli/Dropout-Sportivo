```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore")
```

```python
import pickle
LogisticRegression = pickle.load(open("LogisticRegression FS", "rb"))
RandomForest = pickle.load(open("RandomForest FS", "rb"))
SVM = pickle.load(open("SVM FS", "rb"))
KNN = pickle.load(open("KNN FS", "rb"))
MLP = pickle.load(open("MLP FS", "rb"))
VotingSoft = pickle.load(open("VotingSoft FS", "rb"))
```

```python
acc = np.concatenate((VotingSoft['cv_results'].filter(like="accuracy"),
                      LogisticRegression['cv_results'].filter(like="accuracy"),
                      MLP['cv_results'].filter(like="accuracy"), 
                      SVM['cv_results'].filter(like="accuracy"),
                      KNN['cv_results'].filter(like="accuracy"),
                      RandomForest['cv_results'].filter(like="accuracy")))
data = [acc[0], acc[1], acc[2], acc[3], acc[4], acc[5]]

fig, ax = plt.subplots(figsize=(10,8))
ax.boxplot(data, meanline=True, showmeans=True, meanprops = dict(linestyle='--', linewidth=3))
ax.set_title("Accuracy")
plt.xticks([1, 2, 3, 4, 5, 6], ["VotingSoft", "LogisticRegression", "MLP", "SVM", "KNN", "RandomForest"])
plt.show()
```

```python
dropout = pd.read_csv("04 important_features.csv")
dropout.drop("Unnamed: 0", 1, inplace=True)
X = dropout.drop("dropout", 1)
X["x1"], X["x2"] = X.femmina_pallavolo_nuoto, X.personalita_aperta
X.drop(["femmina_pallavolo_nuoto", "personalita_aperta"], 1, inplace=True)
X["femmina_pallavolo_nuoto"], X["personalita_aperta"] = X["x1"], X["x2"]
X.drop(["x1", "x2"], 1, inplace=True)
y = dropout.dropout
X.head()
```

```python
y_pred_lr = round(pd.DataFrame(LogisticRegression['model'].predict_proba(X)).iloc[:, 1], 2)
y_pred_rf = round(pd.DataFrame(RandomForest['model'].predict_proba(X)).iloc[:, 1], 2)
y_pred_svm = round(pd.DataFrame(SVM['model'].predict_proba(X)).iloc[:, 1], 2)
y_pred_knn = round(pd.DataFrame(KNN['model'].predict_proba(X)).iloc[:, 1], 2)
y_pred_mlp = round(pd.DataFrame(MLP['model'].predict_proba(X)).iloc[:, 1], 2)
y_pred_vs = round(pd.DataFrame(VotingSoft['model'].predict_proba(X)).iloc[:, 1], 2)
```

```python
from sklearn.metrics import roc_curve
ns_probs = [0 for _ in range(len(y))]
ns_fpr, ns_tpr, ns_t = roc_curve(y, ns_probs)
lr_fpr, lr_tpr, lr_t = roc_curve(y, y_pred_lr)
rf_fpr, rf_tpr, rf_t = roc_curve(y, y_pred_rf)
svm_fpr, svm_tpr, svm_t = roc_curve(y, y_pred_svm)
knn_fpr, knn_tpr, knn_t = roc_curve(y, y_pred_knn)
mlp_fpr, mlp_tpr, mlp_t = roc_curve(y, y_pred_mlp)
vs_fpr, vs_tpr, vs_t = roc_curve(y, y_pred_vs)
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(ns_fpr, ns_tpr, linestyle='--', label='No-Skill')
ax.plot(lr_fpr, lr_tpr, marker='.', label='LogisticRegression')
ax.plot(rf_fpr, rf_tpr, marker='.', label='RandomForest')
ax.plot(svm_fpr, svm_tpr, marker='.', label='SVM')
ax.plot(knn_fpr, knn_tpr, marker='.', label='KNN')
ax.plot(mlp_fpr, mlp_tpr, marker='.', label='MLP')
ax.plot(vs_fpr, vs_tpr, marker='.', label='VotingSoft')
ax.set_title("ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
print("AVG AUC VotingSoft: \t    ", VotingSoft['metrics']['mean_test_roc_auc'].iloc[0])
print("AVG AUC MLP: \t\t    ", MLP['metrics']['mean_test_roc_auc'].iloc[0])
print("AVG AUC LogisticRegression: ", LogisticRegression['metrics']['mean_test_roc_auc'].iloc[0])
print("AVG AUC RandomForest: \t    ", RandomForest['metrics']['mean_test_roc_auc'].iloc[0])
print("AVG AUC SVM: \t\t    ", SVM['metrics']['mean_test_roc_auc'].iloc[0])
print("AVG AUC KNN: \t\t    ", KNN['metrics']['mean_test_roc_auc'].iloc[0])
```
